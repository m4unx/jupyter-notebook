# Data Cleaning 

```python
df['SR2'] = df['Specific Risk'].str.lower()
```

```python
df.head()
```

```python
searchfor = ['integrity', 'corruption', 'governance']
df1 = df[df['SR2'].str.contains('|'.join(searchfor), na=False)]
```

```python
df1.shape
```

```python
df1.head()
```

```python
def my_tags():
  condition = [df1['SR2'].str.contains(pat= '^(?=.*corruption)((?!.*corporate).*)', regex=True, case=False), df1['SR2'].str.contains('integrity'), df1['SR2'].str.contains('governance')] #regex to include words corruption and exclude words corporate if exist
  tagname = ['Corruption', 'Integrity', 'Governance']
  df1['tags'] = np.select(condition, tagname)
  return
```

```python
my_tags()
```

```python
df1
```

```python
df1['LOB'] = df1['Risk ID'].str.split(".", 2).str[1] #split column  select second split
```

```python
df1.head()
```

```python
df1['Residual Rating'].unique()
```

```python
Rank_Residual_Rating = {'Low':1, 'Moderate':2, 'Significant':3, 'High':4, 'Extreme':5}
```

```python
df1['#Residual_rating'] = df1['Residual Rating'].map(Rank_Residual_Rating).fillna(0)
```

```python
df1.head()
```

```python
df1.shape
```

```python
df2 = df1.loc[df1['tags']!='0',:]
```

```python
df2 = df2.reset_index()
```
